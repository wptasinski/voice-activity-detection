```python
short, sr = ls.load(short_sound.path)
noise, sr = ls.load(noise_sound.path)
noise_ = noise[1*sr:2*sr]
```

```python
idx = int(np.floor( (sr - len(short))/2 ))
final = np.zeros(sr)
final[idx:idx+len(short)] = short
waveplot(final)
```

```python
noise_ = noise[5*sr:6*sr]
result = (0.5 * final) + (0.03 * noise_)
fig, ax = plt.subplots(figsize=(15,5))
waveplot(result,ax=ax,alpha=0.5)
ax2=ax.twinx()
waveplot(final,alpha=0.5,ax=ax2,color='orange')

ipd.Audio(result,rate=sr)
```
